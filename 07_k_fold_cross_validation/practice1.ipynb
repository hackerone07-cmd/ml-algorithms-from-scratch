```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.svm import SVC
from sklearn import datasets

iris = datasets.load_iris()
```

```python
l_score = cross_val_score(LogisticRegression(max_iter=1000), iris.data, iris.target,cv=3)
np.average(l_score)
```

```python
s_score = cross_val_score(SVC(), iris.data, iris.target,cv=3)
np.average(s_score)
```

```python
d_score = cross_val_score(DecisionTreeClassifier(max_depth=9), iris.data, iris.target,cv=3)
np.average(d_score)
```

```python
r_score = cross_val_score(RandomForestClassifier(n_estimators = 80), iris.data, iris.target,cv=3)
np.average(r_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Example dataset
X = np.array([[1,2],[1,4],[1,0],
              [10,2],[10,4],[10,0]])

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Results
print("Cluster centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)

# Plot
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black', marker='x')
plt.show()
```
