```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
iris = datasets.load_iris()
```

```python
# Create DataFrame with only petal length and width
df = pd.DataFrame(iris.data, columns=iris.feature_names)
petal_df = df[["petal length (cm)", "petal width (cm)"]]
df
```

```python
df = df.drop(["sepal length (cm)","sepal width (cm)"], axis="columns")
```

```python
df
```

```python
plt.scatter(df["petal length (cm)"],df["petal width (cm)"])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
y_predicted = km.fit_predict(df[["petal length (cm)","petal width (cm)"]])
```

```python
df["y_predicted"] = y_predicted
```

```python
df
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df1 = df[df.y_predicted==0]
df2 = df[df.y_predicted==1]
df3 = df[df.y_predicted==2]
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color='green')
plt.scatter(df2["petal length (cm)"],df2["petal width (cm)"],color='red')
plt.scatter(df3["petal length (cm)"],df3["petal width (cm)"],color='pink')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color="purple", marker='+')
plt.xlabel('petel lenght')
plt.ylabel('patel width')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['petal length (cm)']])
df['petal length (cm)'] = scaler.transform(df[['petal length (cm)']])

scaler.fit(df[["petal width (cm)"]])
df["petal width (cm)"] = scaler.transform(df[["petal width (cm)"]])
```

```python
sse = []
k_rng = range(1,8)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[["petal length (cm)","petal width (cm)"]])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
