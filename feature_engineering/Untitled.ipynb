```python
df2 = df.dropna()
```

```python
min_throught, max_throught = df2["price"].quantile([0.1,0.99])
min_throught, max_throught
```

```python
df2[df2["price"] <min_throught]
```

```python
df2[df2["price"] >max_throught]
```

```python
df2 = df2[(df['price']>min_throught) & (df2['price']<max_throught)]
df2.shape
```

```python
df2["z_score"] = (df2.price - df2.price.mean())/df2.price.std()
df2.head(5)
```

```python
df2[df2["z_score"]<4]
```
