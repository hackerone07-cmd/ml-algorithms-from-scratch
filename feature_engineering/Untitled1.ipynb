```python
df.drop(["id","host_id","host_name",	"neighbourhood_group",	"neighbourhood",	"latitude",	"longitude","number_of_reviews","last_review",	"reviews_per_month",	"calculated_host_listings_count"], axis="columns",inplace=True)
```

```python
df.isna().sum()
```

```python
df2 = df.copy()
df2 = df.dropna()
```

```python
df2.isna().sum()
```

```python
df2["price_per_night"] = df2["price"]/df2["minimum_nights"]
```

```python
df2
```

```python
min_throught, max_throught = df2["price_per_night"].quantile([0.8,0.999])
min_throught, max_throught
```

```python
df2[df2["price_per_night"] <min_throught]
```

```python
df2[df2["price_per_night"] >max_throught]
```

```python
df3 = df2.copy()
df3 = df2[(df2['price_per_night']>min_throught) & (df2['price_per_night']<max_throught)]
df3
```

```python
df3.describe()
```

```python
upper = df3.price_per_night.mean() + 4* df3.price_per_night.std()
lower = max(0,df3.price_per_night.mean() - 4* df3.price_per_night.std())
upper,lower
```

```python
df3[(df3.price_per_night>upper) | (df3.price_per_night<lower) ]
```

```python
df3  = df3[(df3.price_per_night<upper) & (df3.price_per_night>lower)]
df3
```
