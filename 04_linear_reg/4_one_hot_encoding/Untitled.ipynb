```python
import pandas as pd
```

```python
df= pd.read_csv("carprices.csv")
df
```

```python
dummy = pd.get_dummies(df["Car Model"]).astype(int)
dummy
```

```python
merge = pd.concat([df,dummy], axis="columns")
merge
```

```python
final = merge.drop(["Car Model"], axis="columns")
final
```

```python
new_final = final.drop(["Mercedez Benz C class"], axis="columns")
new_final
```

```python
X= new_final.drop(["Sell Price($)"], axis="columns")
X
```

```python
y = df["Sell Price($)"]
y
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```
