```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.scatter(df.year,df["per capita income (US$)"], color="red", marker="+")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
```

```python
X = df[["year"]]   # feature
y = df["per capita income (US$)"]   # target
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
predicted_income_2020 = model.predict([[2020]])
print("Predicted income in 2020:", predicted_income_2020)
```

```python
plt.scatter(X, y, color="blue")   # actual data
plt.plot(X, model.predict(X), color="red")   # regression line
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```
