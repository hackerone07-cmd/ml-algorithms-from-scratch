```python
import pandas as pd
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales
```

## Line Chart

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="yellow", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales month 1")
plt.xlabel("Quarter->")
plt.ylabel("Million $")

plt.legend()

plt.show()
```

## Pie Chart

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=120, explode=(0.1,0,0), shadow=True)
plt.show()
```

## Bar Chart

```python
df_sales
```

```python
df_sales.plot(kind="bar", x="Quarter")

plt.xticks(rotation=50)
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2= df_sales.set_index("Quarter")
df_sales_2
```

```python
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="scatter",x="Dishwasher", y="Fridge")
plt.show()
```
